```
# Import Libraries
import numpy as np
import pandas as pd

import yfinance as yf

# 
import plotly.graph_objs as go
```

```
df = yf.download(tickers='GAIL.NS', period='1d', interval='1m')


# Calculate the middel band and lower band
df['middle'] = df['Close'].rolling(window=20).mean()
df['upper'] = df['middle'] + 1.96 * df['Close'].rolling(window=21).std()
df['lower'] = df['middle'] - 1.96 * df['Close'].rolling(window=21).std()

# declare a figure
fig=go.Figure()

fig.add_trace(go.Scatter(x=df.index, y=df['middle'], line=dict(width=1.5, color = 'blue'), name = 'Middle Band'))
fig.add_trace(go.Scatter(x=df.index, y=df['lower'], line=dict(width=1.5, color = 'red'), name = 'Lower Band'))
fig.add_trace(go.Scatter(x=df.index, y=df['upper'], line=dict(width=1.5, color = 'green'), name = 'Upper Band'))

# Candle Sticks
fig.add_trace(go.Candlestick(x=df.index, open=df['Open'], high=df['High'], low=df['Low'], close=df['Close'], name = 'Market Data'))

# Add titles
fig.update_layout(title = 'Inox Live share price evolution', yaxis_title='Stock Price per INR')

# X-Axes
fig.update_xaxes(rangeslider_visible=True, rangeselector=dict(buttons=list([dict(count=15, label='15m', step='minute', stepmode='backward'),
                                                                            dict(count=45, label='45m', step='minute', stepmode='backward'),
                                                                            dict(count=1, label='HTD', step='hour', stepmode='todate'),
                                                                            dict(count=1, label='Day', step='day', stepmode='todate'),
                                                                            dict(step='all')])))
# Show

fig.show()
```
