```
import pandas as pd
import numpy as np
import pandas_datareader as web
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

import plotly.graph_objects as go
```

```
# Get the Stock Symbol for the Portfolio 
# FAANG
stockSymbols = ['FB']

# GET the stock starting date
stockDate = '2019-01-01'

stock_End_date = '2020-06-30'
```

```
# Create a function to get the stock prices in the portfolio
def getMyPortfolio(stocks = stockSymbols, start = stockDate, end=stock_End_date):
  data = web.DataReader(stocks, data_source='yahoo', start=start, end=end, )
  data.columns = ['AdjClose', 'Close', 'High', 'Low', 'Open', 'Volume']
  return data
```

```
# Get the stock portfolio Adj. Close price
my_stocks = getMyPortfolio(stockSymbols)
my_stocks
```

```
# Get the Highest price from the dataset
max_price = my_stocks.Close.max()

# Get the lowest price from the dataset
low_price = my_stocks.Close.min()

print(max_price)
print(low_price)
```

```
# Fibonacci Levels considering original trend as upward move
diff = max_price - low_price
level1 = max_price - 0.236 * diff
level2 = max_price - 0.382 * diff
level3 = max_price - 0.618 * diff

print("Level :", "Price")
print("0     :", max_price)
print("0.236 :", (level1))
print("0.382 :", level2)
print("0.618 :", level3)
print("1     :", low_price)
```

```
# declare a figure
fig=go.Figure()

fig.add_trace(go.Candlestick(x=my_stocks.index, open=my_stocks['Open'], high=my_stocks['High'], low=my_stocks['Low'], close=my_stocks['Close'], name = 'Market Data'))

fig.update_layout(shapes=(list([dict(type= 'line', yref= 'y', y0= level1, y1= level1, xref= 'paper', x0= 0, x1= 1),
                          dict(type= 'line', yref= 'y', y0= level2, y1= level2, xref= 'paper', x0= 0, x1= 1),
                          dict(type= 'line', yref= 'y', y0= level3, y1= level3, xref= 'paper', x0= 0, x1= 1),
                          dict(type= 'line', yref= 'y', y0= max_price, y1= max_price, xref= 'paper', x0= 0, x1= 1),
                          dict(type= 'line', yref= 'y', y0= low_price, y1= low_price, xref= 'paper', x0= 0, x1= 1)])))

fig.show()
```
